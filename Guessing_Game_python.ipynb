```python
import random

jackpot = random.randint(1, 100)
counter = 0

print("🎯 Guess the number between 1 and 100")

while True:
    try:
        guess = int(input("Enter your number: "))
        counter += 1

        if guess < jackpot:
            print("📉 Guess Higher")
        elif guess > jackpot:
            print("📈 Guess Lower")
        else:
            print("✅ Correct Guess!")
            print(f"You took {counter} attempts")
            break

    except ValueError:
        print("❌ Please enter a valid number!")
```
